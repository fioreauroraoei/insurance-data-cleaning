```python
from scipy.stats import skew

# Hitung skewness untuk age dan claim_amount
for col in ['age', 'claim_amount']:
    if col in df.columns:
        skewness = df[col].skew()
        print(f"Skewness of {col}: {skewness}")

        if abs(skewness) < 0.5:
            # distribusi cukup simetris → pakai mean
            df[col] = df[col].fillna(df[col].mean())
            print(f"Imputed {col} with mean")
        else:
            # distribusi miring → pakai median
            df[col] = df[col].fillna(df[col].median())
            print(f"Imputed {col} with median")
```

```python
# Impute age with median
if 'age' in df.columns:
    df['age'] = df['age'].fillna(df['age'].median())
    
# Impute claim_amount with your choice (mean or median) – default median here
if 'claim_amount' in df.columns:
    df['claim_amount'] = df['claim_amount'].fillna(df['claim_amount'].median())

# Impute diagnosis_code with mode (most frequent)
if 'diagnosis_code' in df.columns:
    mode_val = df['diagnosis_code'].mode(dropna=True)
    if len(mode_val) > 0:
        df['diagnosis_code'] = df['diagnosis_code'].fillna(mode_val.iloc[0])

missing_after = df.isna().sum()
print('\nMissing after:\n', missing_after)
```

```python
# Plot BEFORE
if 'claim_amount' in df.columns:
    plt.figure()
    plt.hist(df['claim_amount'].dropna(), bins=30)
    plt.title('Claim Amount – Before')
    plt.xlabel('claim_amount'); plt.ylabel('count')
    plt.show()

    # IQR method
    q1 = df['claim_amount'].quantile(0.25)
    q3 = df['claim_amount'].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr
    print(f'IQR bounds: [{lower:.2f}, {upper:.2f}]')

    # Example: capping (winsorization)
    df['claim_amount_capped'] = df['claim_amount'].clip(lower, upper)

    # Plot AFTER
    plt.figure()
    plt.hist(df['claim_amount_capped'].dropna(), bins=30)
    plt.title('Claim Amount – After Capping')
    plt.xlabel('claim_amount_capped'); plt.ylabel('count')
    plt.show()
else:
    print("Column 'claim_amount' not found; skipping outlier section.")
```

- Create a boxplot of claim_amount.
The initial boxplot of `claim_amount` shows several extreme values above the upper whisker, indicating potential outliers.

- Identify potential outliers using the IQR method.
Q1 = X
Q3 = Y
IQR = Q3 - Q1
Lower bound = Q1 - 1.5 x IQR
Upper bound = Q3 + 1.5 x IQR
Any values outside these bounds are considered potential outliers.
Lower bound ≈ -3257.72 (ignore, since clain amount cannot be negative)
Upper bound ≈ 10532.82 (all claims above 10532.82 are flagged as outliers)

- Decide whether to remove, cap, or transform them (justify your approach).
I decide to cap the outliers (winsorization)
Reasoning:
Removing rows may may result in loss of valuable information; claim amounts are naturally skewes (long tail distribution), so extreme but valid cases are possible; Capping ensures extreme values do not overly influence modeling while keeping the data consistent.

- Show before-and-after distribution plots.
Before: Right-skewed distribution with very high extreme values (up to >20,000)
After Capping: Distribution is still right-skewed but extreme values are capped at -10,532, reducing the influence of outliers.

```python
from sklearn.preprocessing import StandardScaler
# ambil kolom kategorikal yang ada
categoricals = [c for c in ['gender', 'policy_type', 'payment_status', 'diagnosis_code'] if c in df.columns]
df_encoded = pd.get_dummies(df, columns=categoricals, drop_first=True)
# scaling claim_amount atau claim_amount_capped
target_col = 'claim_amount_capped' if 'claim_amount_capped' in df_encoded.columns else 'claim_amount'
if target_col in df_encoded.columns:
    scaler = StandardScaler()
    df_encoded[target_col + '_z'] = scaler.fit_transform(df_encoded[[target_col]])
else:
    print("No claim amount column available for scaling.")
#print bentuk data setelah encoding dan scaling
print('Encoded columns added. New shape:', df_encoded.shape)
df_encoded.head()
```

karena variabel kategorikal seperti itu tidak bisa langsung dipakai di model ML, perlu diencode dulu. Kita pakai One-Hot Encoding dengan membuat colom dummy


- Normalize or standardize the claim_amount variable.

Supaya semua fitur punya skala yang sebanding. Algoritma berbasis jarak (KNN, SVM, Clustering) butuh skala beragam supaya hasil akurat.
